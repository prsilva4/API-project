```python
import requests
import pandas as pd
import json
```

```python
stl = []
dreb = []
first_name=[]
last_name=[]
season=[]

for i in range(331):

    url = "https://free-nba.p.rapidapi.com/stats?seasons[]=2018"

    querystring = {"page":i,"per_page":"100"}

    headers = {'x-rapidapi-host': "free-nba.p.rapidapi.com",
               'x-rapidapi-key': "0a58c448e1msh42682201a62cc30p1ea6c2jsnb4992c5a1579"}

    result = requests.get(url, headers=headers, params=querystring)
    
    data = result.json()
    
    for e in range(len(data['data'])):
        
        stl.append(data['data'][e]['stl'])
        
        dreb.append(data['data'][e]['dreb'])
        
        first_name.append(data['data'][e]['player']['first_name'])
        
        last_name.append(data['data'][e]['player']['last_name'])
        
        season.append(data['data'][e]['game']['season'])

        
```

```python
df = pd.DataFrame(list(zip(first_name, last_name,stl,dreb)),
               columns =['First_Name', 'Last_Name','stl','dreb'])

df['Player']=df['First_Name']+' '+ df['Last_Name']

df=df[['Player','stl','dreb']]

df=df.groupby(by='Player').mean()

df['season']='2018'

df
```
