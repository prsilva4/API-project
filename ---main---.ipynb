```python
stl = []
dreb = []
first_name=[]
last_name=[]

for i in range(318):

    url = "https://free-nba.p.rapidapi.com/stats?seasons[]=2020"

    querystring = {"page":i,"per_page":"100"}

    headers = {'x-rapidapi-host': "free-nba.p.rapidapi.com",
               'x-rapidapi-key': "0a58c448e1msh42682201a62cc30p1ea6c2jsnb4992c5a1579"}

    result = requests.get(url, headers=headers, params=querystring)
    
    data = result.json()
    
    
    for i in range(len(data['data'])):
        
        stl.append(data['data'][i]['stl'])
        
        dreb.append(data['data'][i]['dreb'])
        
        first_name.append(data['data'][i]['player']['first_name'])
        
        last_name.append(data['data'][i]['player']['last_name'])
```

```python
df = pd.DataFrame(list(zip(first_name, last_name,stl,dreb)),
               columns =['First_Name', 'Last_Name','stl','dreb'])
```

```python
df['Player']=df['First_Name']+' '+ df['Last_Name']
```

```python
df=df[['Player','stl','dreb']]
```

```python
df=df.groupby(by='Player').mean()

#df['season']='2020'

df
```

```python
df.reset_index(inplace=True)
```
